```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
sns.set()
#
# Load the insurance3r2.csv data set
#
claims_full_data = pd.read_csv('insurance3r2.csv')
#
#
claims = claims_full_data.sample(n=100, random_state=100)
claims.sort_values("age", inplace=True)
sns.relplot(x='age', y='charges', col='sex', hue='region', data=claims)
```

```python
sns.catplot(x='region', y='charges', hue='sex', kind='swarm', data=claims)
```

```python
sns.lmplot(x='charges', y='age', col='region', hue='sex', data=claims)
```

```python
sns.catplot(x='region', y='charges', hue='sex', kind='violin', data=claims)
```

```python
x = claims['age']
sns.distplot(x)
```
